```python
%matplotlib inline
from torch.autograd import Variable
import numpy as np
import torch
torch.cuda.set_device(1)
from dataset import *
from torch.utils.data import DataLoader
from torch.utils.data.sampler import RandomSampler
import torchsample
from matplotlib import pyplot as plt
def showTorchImage(im, figsize=(4,4), vmin=-1.0, vmax=1.0, cmap='gray'):
    if type(im) == np.ndarray:
        im = im
    elif type(im) == list:
        im = numpy.array(im)
    else:
        im = im.cpu().data.numpy()
        
    image_size = im.shape[-1]
    im = im.reshape((-1, image_size, image_size))
    n = int(np.sqrt(im.shape[0])) ** 2
    im = im[:n].reshape((int(np.sqrt(n)), int(np.sqrt(n)), image_size, image_size))
    im = im.transpose(0,2,1,3).reshape((int(np.sqrt(n))*image_size, int(np.sqrt(n))*image_size))
    plt.figure(figsize = figsize)
    plt.imshow(im, cmap=cmap, vmin=vmin, vmax=vmax)
class InfiniteRandomSampler(RandomSampler):

    def __iter__(self):
        while True:
            it = super().__iter__()
            for x in it:
                yield x
```
